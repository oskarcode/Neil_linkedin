```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
connection_df = pd.read_csv('./Connections.csv')
connection_df.info()
```

```python
import plotly.io as pio
import plotly.express as px
pio.renderers
pio.renderers.default = "notebook_connected"
time = px.line(connection_df.groupby(by='Connected On').count().reset_index(),
x = 'Connected On',
y = 'First Name',
labels = {'First Name':'Number of Connections'},
title = 'Connection by time')
time.show()
```

```python
import datetime

def convert(date):
          return datetime.datetime.strptime(date, "%d %b %Y").strftime("%Y-%m-%d")

connection_df["Connected On"] = connection_df["Connected On"].apply(convert)
connections_line = px.line(connection_df.groupby(by='Connected On').count().reset_index(), 
                           x="Connected On", 
                           y="First Name", 
                           labels={'First Name': 'Number'},
                           title='My Connections')
connections_line.show()
```

```python
company_group = connection_df.groupby(by='Company').count().reset_index()
company_group = company_group.sort_values(by = 'Connected On', ascending = False)

fig = px.bar(company_group[:20],
x = 'Company',
y = 'Connected On',
labels = {'Connected On': 'Number of Connections'},
width = 1000,
height = 900,
title = 'Connections by company')
fig.show()
```

```python
fig2 = px.treemap(company_group[:200],path=['Company','Position'],
values = 'Connected On',
labels= {'Connedted On': 'Number of Connections'},
width=1000,
height = 900,
title='Tree map for connection by company')
fig2.show()
```

```python
fig3 = px.bar(connection_df.groupby(by='Position').count().reset_index().sort_values(by='First Name', ascending=False)[:10],
x = 'Position',
y ='Connected On',
labels={'Connected On': 'Number of Connections'},
title = 'Position by connections')
fig3.show()
```

```python
position_group = connection_df.groupby(by='Position').count().reset_index()
company_position_group = position_group.sort_values(by = 'Connected On', ascending = False)



fig3 = px.treemap(company_position_group[:200],path=['Company','Position'],
values = 'Connected On',
labels= {'Connedted On': 'Number of Connections'},
width=1000,
height = 900,
title='Tree map for connection by company')
fig3.show()
```
